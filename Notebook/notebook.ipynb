```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text) :
    L = []
    for i in ast.literal_eval(text) :
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords'][0]
```

```python
def convert_top_3(text) :
    counter = 0
    L = []
    for i in ast.literal_eval(text) :
        if(counter<3) :
            L.append(i['name'])
        counter=counter+1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_top_3)
movies['cast'][0]
```

```python
def fetch_director(text) :
    L = []
    for i in ast.literal_eval(text) :
        if i['job'] == 'Director' :
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew'][345]
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
movies['tagline'] = movies['tagline'].apply(lambda x : x.split())
movies.sample(1)
```

```python
def collapse_space(L) :
    L1 = []
    for i in L :
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(collapse_space)
movies['keywords'] = movies['keywords'].apply(collapse_space)
movies['cast'] = movies['cast'].apply(collapse_space)
movies['crew'] = movies['crew'].apply(collapse_space)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] +movies['tagline'] +movies['cast']+movies['crew']+movies['keywords']
```

```python
new_movies = movies.drop(columns = ['overview', 'cast', 'crew', 'keywords', 'tagline', 'genres'])
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(x))
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new_movies.head(1)
```

```python
new_movies[new_movies['title']=='The Lego Movie'].index[0]
```

```python
print(similarity)
```

```python
def recommend(movie_title) :
    index = new_movies[new_movies['title']==movie_title].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    for i in distances[1:10] :
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Fantastic Four')
```
